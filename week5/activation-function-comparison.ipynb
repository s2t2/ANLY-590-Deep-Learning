```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from __future__ import division
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return np.maximum(x,0)

def tan_h(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.arange(-2.0, 2.0, 0.01)
```

```python
sg = sigmoid(x)
plt.plot(x, sg)
```

```python
rl = relu(x)
plt.ylim(0,1)
plt.plot(x, rl)
```

```python
th = tan_h(x)
plt.plot(x, th)
```

```python
plt.figure(figsize=(15,10))
plt.plot(x, sg)
plt.plot(x, rl)
plt.plot(x, th)
plt.show()
```
